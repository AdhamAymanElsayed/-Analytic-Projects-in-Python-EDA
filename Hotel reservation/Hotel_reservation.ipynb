### 1 -  Number of booking for each number of adults with each number of children 

```python
df.groupby(['no_of_adults','no_of_children'])['Booking_ID'].count().reset_index()
```

### 2 - the highest meal are ordered

```python
df.type_of_meal_plan.value_counts().sort_values(ascending = False).reset_index()
```

### 3-Required car parking space 

```python
df.groupby('required_car_parking_space')['Booking_ID'].count().sort_values(ascending = False).reset_index()
```

## or

```python
df['required_car_parking_space'].value_counts()
```

### 4 - Gain of money for each room

```python
df.groupby('room_type_reserved')['avg_price_per_room'].sum().sort_values(ascending = False).reset_index()
```

### 5 - number of special requests

```python
df.groupby('no_of_special_requests')['Booking_ID'].count().sort_values(ascending = False).reset_index()
```

### 6 - Number of booking that canceled and that not canceled

```python
df.booking_status.value_counts()
```

```python
df.corr()
```

### 7 - Week nights

```python
df.groupby('no_of_week_nights')['Booking_ID'].count().sort_values(ascending = False).reset_index()
```
