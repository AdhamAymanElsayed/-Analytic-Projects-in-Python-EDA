Ask yourself 15 analystical questions that you belive will help the managers in decision making

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'ecommerce.csv')
df.head()
```

### 1- What are the total Order value for each country


```python
x = df.groupby('Country')['OrderValue'].sum()
x.to_frame()
```

### 2- What each Major Category sold quantity during November 2011


```python
df['Year-Month'] = pd.to_datetime(df['Year-Month'])
```

```python
df.info()
```

```python
df['Month']  = pd.DatetimeIndex(df['Year-Month']).month
df['Year']  = pd.DatetimeIndex(df['Year-Month']).year
```

```python
df.head(2)
```

```python
df[(df['Month'] == 11) & (df['Year'] == 2011)].groupby('Major Category')['Quantit'].sum().reset_index()
```

### 3- What's Total Order Value for each Minor Item in Kitchen


```python
df.groupby(['Major Category','Minor Category'])['OrderValue'].sum().reset_index()
```

### 4- What Month/s contributes to high demand in orders

```python
df.groupby('Year-Month')['Quantit'].sum().sort_values(ascending = False).reset_index()
```

```python
df.groupby('Year-Month')['Quantit'].sum().max()
```

### 5- What Month/s contributes to high OrderValue?

```python
df.groupby('Year-Month')['OrderValue'].sum().sort_values(ascending = False).reset_index()
```

```python
df.groupby('Year-Month')['OrderValue'].sum().max()
```

### 6- Why April and December suffer from low sales? 


```python
df[(df['Month'] == 4) | (df['Month'] == 12)].groupby(['Major Category','Minor Category','Country'])['OrderValue'].sum().reset_index()
```

### 7- Why May and November have high sales? 


```python
df[(df['Month'] == 5) | (df['Month'] == 11)].groupby(['Major Category','Minor Category','Country'])['OrderValue'].sum().reset_index()
```

### 8- Most Major Items sold in April and December


```python
df[(df['Month'] == 4) | (df['Month'] == 12)].groupby('Major Category')['Quantit'].sum().sort_values(ascending = False ).to_frame()
```

### 9- Most Major Items sold in May and November


```python
df[(df['Month'] == 5) | (df['Month'] == 11)].groupby('Major Category')['Quantit'].sum().sort_values(ascending = False ).to_frame()
```

### 10- What Hour recives high demand for ordering?


```python
df['hour'] = df['InvoiceTime'].str.split(':').str[0]
df['hour']
```

```python
df.groupby('hour')['Quantit'].sum().reset_index()
```

### 11 - Order VAlue For each Country in each month


```python
df.pivot_table(index = 'Year-Month',values = 'OrderValue',aggfunc='sum',columns= 'Country')
```

### 12 - Order VAlue For each Country and Major Category in Minor Category


```python
df.pivot_table(index = 'Year-Month',values = 'OrderValue',aggfunc='sum',columns= ['Country','Major Category']).T.reset_index()
```

### 13 - Order VAlue For each Country and Major Category in Minor Category

```python
df.pivot_table(index = 'hour',values = 'OrderValue',aggfunc='sum',columns= ['Country','Major Category']).T.reset_index()
```

### 14- Correlation between 'Quantit','OrderValue','UnitPrice'.


```python
df[['Quantit','OrderValue','UnitPrice']].corr()[['Quantit','OrderValue','UnitPrice']]
```
